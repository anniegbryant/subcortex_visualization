```python
import matplotlib.colors as mcolors

# 1. aseg 
white_to_purple_cmap = mcolors.LinearSegmentedColormap.from_list("white_to_purple", ["white", "purple"])
aseg_demo = plot_subcortical_data(atlas='aseg', hemisphere='L', 
                      fill_title="aseg atlas region", 
                      show_figure=False,
                      cmap=white_to_purple_cmap)

# 2. Melbourne Subcortical Atlas, S1
white_to_blue_cmap = mcolors.LinearSegmentedColormap.from_list("white_to_blue", ["white", "#1D3B9F"])
tian_S1_demo = plot_subcortical_data(atlas='Melbourne_S1', hemisphere='L', 
                      fill_title="Melbourne Subcortical atlas (S1) region",
                      show_figure=False,
                      cmap=white_to_blue_cmap)

# 3. Melbourne Subcortical Atlas, S2
white_to_green_cmap = mcolors.LinearSegmentedColormap.from_list("white_to_green", ["white", "#7FB343", "#253F25"])
tian_S2_demo = plot_subcortical_data(atlas='Melbourne_S2', hemisphere='L', 
                      fill_title="Melbourne Subcortical atlas (S2) region", 
                      show_figure=False,
                      cmap=white_to_green_cmap)

# 4. Melbourne Subcortical Atlas, S3
# transition between green and blueteal region
white_to_teal_cmap = mcolors.LinearSegmentedColormap.from_list(
    "white_to_teal", ["white", "#00A6A6", "#005F73"]
)
tian_S3_demo = plot_subcortical_data(atlas='Melbourne_S3', hemisphere='L', 
                      fill_title="Melbourne Subcortical atlas (S3) region", 
                      show_figure=False,
                      cmap=white_to_teal_cmap)

# 5. Melbourne Subcortical Atlas, S4
# transition continuing toward cyanturquoise before yellow
white_to_turquoise_cmap = mcolors.LinearSegmentedColormap.from_list(
    "white_to_turquoise", ["white", "#5CE1E6", "#00BFB3"]
)
tian_S4_demo = plot_subcortical_data(atlas='Melbourne_S4', hemisphere='L', 
                      fill_title="Melbourne Subcortical atlas (S4) region", 
                      show_figure=False,
                      cmap=white_to_turquoise_cmap)

# 6. AICHA
white_to_yellow_cmap = mcolors.LinearSegmentedColormap.from_list("white_to_yellow", ["white", "#FAE055", "#EF8036"])
AICHA_demo = plot_subcortical_data(atlas='AICHA', hemisphere='L', 
                      fill_title="AICHA Subcortical atlas region", 
                      show_figure=False,
                      cmap=white_to_yellow_cmap)

# 7. Brainnetome
white_to_red_cmap = mcolors.LinearSegmentedColormap.from_list("white_to_red", ["white", "#ED7570", "#C73125"])
brainnetome_demo = plot_subcortical_data(atlas='Brainnetome', hemisphere='L', 
                      fill_title="Brainnetome Subcortical atlas region", 
                      show_figure=False,
                      cmap=white_to_red_cmap)

# 8. Thalamic nuclei HCP (Najdenovska 2018)
white_to_pink_cmap = mcolors.LinearSegmentedColormap.from_list("white_to_pink", ["white", "#DE94B9", "#B23167"])
thalamus_demo = plot_subcortical_data(atlas='Thalamus_Nuclei_HCP', hemisphere='L', 
                      fill_title="Thalamic nuclei (Najdenovska 2018) atlas region", 
                      show_figure=False,
                      cmap=white_to_pink_cmap)

# 9. SUIT
# final step into purplish-pink, linking redpink end of the range
white_to_magenta_cmap = mcolors.LinearSegmentedColormap.from_list(
    "white_to_magenta", ["white", "#E48DFB", "#C04BF7"]
)
SUIT_demo = plot_subcortical_data(atlas='SUIT', hemisphere='both', 
                      fill_title="SUIT Subcortical atlas region", 
                      show_figure=False,
                      cmap=white_to_magenta_cmap)
```
